```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("casasboston.csv")
```

```python
print(datos)
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]
```

```python
df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS": "RIO CHARLES",
    "RM": "N_HABITACIONES_MEDIO",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT": "PCT_CLASE_BAJA"
})
```

```python
print(df)
```

```python
df.N_HABITACIONES_MEDIO.plot.hist(color='blue', edgecolor='black')
plt.title('Distribución del Número Medio de Habitaciones')
plt.xlabel('Número Medio de Habitaciones')
plt.ylabel('Frecuencia')
plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN",alpha=1, color='blue')
plt.title('Relacion de valor mediano e índice de criminalidad')
plt.xlabel('Valor mediano')
plt.ylabel('Indice de crimen')
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
```

```python
valor_por_ciudad.head(10).plot.barh(color='blue')
plt.title('Promedio de las primeras 10 ciudades')
plt.xlabel('Promedio')
plt.ylabel('Ciudades')
plt.show()
```
